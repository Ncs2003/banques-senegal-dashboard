```python
# ============================================================
# — Dictionnaire et fonctions d'extraction
# ============================================================
BANQUES_SENEGAL = {
    # PDF 2021
    'SGSN': 'SGBS',
    'BICIS': 'BICIS',
    'CBAO': 'CBAO',
    'CDS': 'CDS',
    'BHS': 'BHS',
    'CITIBANK': 'CITIBANK',
    'LBA': 'LBA',
    'BIS': 'BIS',
    'ECOBANK': 'ECOBANK',
    'BOA': 'BOA',
    'BSIC': 'BSIC',
    'BA-Sénégal': 'BAS',
    'BRM': 'BRM',
    'FBNBank': 'FBNBANK',
    'CREDIT INTERNATIONAL': 'CISA',
    'BNDE': 'BNDE',
    'NSIA Banque': 'NSIA Banque',
    'BDK': 'BDK',
    'BGFIBANK': 'BGFI',
    'BCIM': 'BCIM',
    'LBO': 'LBO',
    'CBI-Sénégal': 'CBI',
    'UBA-Sénégal': 'UBA',
    # PDF 2022
    'C.D.S.': 'CDS',
    'B.H.S.': 'BHS',
    'B.I.S.': 'BIS',
    'B.A.-S.': 'BAS',
    'B.R.M.': 'BRM',
    'U.B.A.': 'UBA',
    'B.N.D.E': 'BNDE',
    'NSIA BANQUE': 'NSIA Banque',
    'BGFI BANK': 'BGFI',
    'CBI-SENEGAL': 'CBI',
    'BOA-S': 'BOA',
    'ORABANK': 'ORABANK',
    'FBNBANK': 'FBNBANK',
}

import re

def nettoyer_nombre(val):
    if val is None:
        return None
    val = str(val).replace(' ', '').replace('\n', '').strip()
    try:
        return float(val)
    except:
        return None

def extraire_compte_resultat(page, annee):
    tables = page.extract_tables()
    if not tables:
        return None
    table = tables[0]
    if not table:
        return None

    indicateurs = [
        'INTERETS.ET.PRODUITS.ASSIMILES', 'NTERETS.ET.CHARGES.ASSIMILEES',
        'REVENUS.DES.TITRES.A.REVENU.VARIABLE', 'COMMISSIONS.(PRODUITS)',
        'COMMISSIONS.(CHARGES)', 'GAINS.OU.PERTES.NETS.SUR.OPERATIONS.DES.PORTEFEUILLES.DE.NEGOCIATION',
        'GAINS.OU.PERTES.NETS.SUR.OPERATIONS.DES.PORTEFEUILLES.DE.PLACEMENT.ET.ASSIMILES',
        "AUTRES.PRODUITS.D'EXPLOITATION.BANCAIRE", "AUTRES.CHARGES.D'EXPLOITATION.BANCAIRE",
        'PRODUIT.NET.BANCAIRE', "SUBVENTIONS.D'INVESTISSEMENT", "CHARGES.GENERALES.D'EXPLOITATION",
        'DOTATIONS.AUX.AMORTISSEMENTS.ET.AUX.DEPRECIATIONS.DES.IMMOBILISATIONS.INCORPORELLES.ET.CORPORELLES',
        "RESULTAT.BRUT.D'EXPLOITATION", 'COÛT.DU.RISQUE', "RESULTAT.D'EXPLOITATION",
        'GAINS.OU.PERTES.NETS.SUR.ACTIFS.IMMOBILISES', 'RESULTAT.AVANT.IMPÔT',
        'IMPÔTS.SUR.LES.BENEFICES', 'RESULTAT.NET'
    ]

    # FORMAT 2b
    try:
        for row in table:
            for cell in row:
                if cell and '\n' in str(cell) and 'Millions' in str(cell):
                    lignes = cell.split('\n')
                    lignes = [l for l in lignes if l.strip()
                              and "Millions" not in l
                              and l.strip() not in ['2019', '2020', '2021', '2022', '2023']]
                    if len(lignes) >= 20:
                        data = {}
                        for i, ind in enumerate(indicateurs):
                            if i < len(lignes):
                                mots = lignes[i].strip().split()
                                if len(mots) == 6:
                                    derniere_val = mots[4] + mots[5]
                                elif len(mots) == 3:
                                    derniere_val = mots[2]
                                elif len(mots) == 4:
                                    derniere_val = mots[2] + mots[3]
                                elif len(mots) == 9:
                                    derniere_val = mots[6] + mots[7] + mots[8]
                                elif len(mots) == 2:
                                    derniere_val = mots[1]
                                elif len(mots) == 1:
                                    derniere_val = mots[0]
                                else:
                                    derniere_val = mots[-1]
                                data[ind] = nettoyer_nombre(derniere_val) if derniere_val else None
                            else:
                                data[ind] = None
                        return data
    except:
        pass

    # FORMAT 4
    if (len(table) > 5 and len(table[3]) >= 4
            and table[3][3] and str(table[3][3]).strip().isdigit()
            and table[4] and table[4][1]
            and '\n' not in str(table[4][1])):
        try:
            data = {}
            ind_index = 0
            IGNORER = ['SENEGAL', 'COMPTES', 'PRODUITS/CHARGES', 'MILLIONS', 'SGSN', 'BICIS',
                       'CBAO', 'CDS', 'BHS', 'CITIBANK', 'LBA', 'BIS', 'ECOBANK', 'ORABANK',
                       'BOA', 'BSIC', 'BAS', 'BRM', 'UBA', 'FBNBANK', 'CISA', 'BNDE',
                       'NSIA', 'BDK', 'BGFI', 'LBO', 'CBI', 'BIMAO', 'CREDIT INTERNATIONAL']
            for row in table:
                if not row[0] or not str(row[0]).strip():
                    continue
                if any(x in str(row[0]).upper() for x in IGNORER):
                    continue
                derniere_val = None
                for col in reversed(row):
                    if col is not None and str(col).strip() != '':
                        derniere_val = col
                        break
                if derniere_val and ind_index < len(indicateurs):
                    valeurs = str(derniere_val).split('\n')
                    for v in valeurs:
                        if ind_index < len(indicateurs):
                            data[indicateurs[ind_index]] = nettoyer_nombre(v.strip())
                            ind_index += 1
            if data:
                return data
        except:
            pass

    # FORMAT 2
    try:
        cellule = table[1][1]
        if cellule and "Montant net" in str(cellule):
            lignes = cellule.split('\n')
            lignes = [l for l in lignes if l.strip()
                      and "Montant net" not in l
                      and l.strip() not in ['2019', '2020', '2021', '2022', '2023']]
            data = {}
            for i, ind in enumerate(indicateurs):
                if i < len(lignes):
                    mots = lignes[i].strip().split()
                    if len(mots) == 6:
                        derniere_val = mots[4] + mots[5]
                    elif len(mots) == 3:
                        derniere_val = mots[2]
                    elif len(mots) == 4:
                        derniere_val = mots[2] + mots[3]
                    elif len(mots) == 9:
                        derniere_val = mots[6] + mots[7] + mots[8]
                    elif len(mots) == 2:
                        derniere_val = mots[1]
                    elif len(mots) == 1:
                        derniere_val = mots[0]
                    else:
                        derniere_val = mots[-1]
                    data[ind] = nettoyer_nombre(derniere_val) if derniere_val else None
                else:
                    data[ind] = None
            return data
    except:
        pass

    # FORMAT 3
    for row in table:
        if row[0] and "Intérêts" in str(row[0]):
            derniere_col = None
            for col in reversed(row):
                if col is not None and str(col).strip() != '':
                    derniere_col = col
                    break
            if derniere_col:
                valeurs = derniere_col.split('\n')
                data = {}
                for i, ind in enumerate(indicateurs):
                    data[ind] = nettoyer_nombre(valeurs[i]) if i < len(valeurs) else None
                return data

    # FORMAT 1
    for row in table:
        if row[0] and "Intérêts" in str(row[0]):
            valeurs_col = None
            for col in reversed(row):
                if col is not None and str(col).strip() != '':
                    valeurs_col = col
                    break
            if valeurs_col:
                valeurs = valeurs_col.split('\n')
                data = {}
                for i, ind in enumerate(indicateurs):
                    data[ind] = nettoyer_nombre(valeurs[i]) if i < len(valeurs) else None
                return data

    return None

print("✅ Fonctions définies !")
```

```python
# ============================================================
# — Extraction PDF 2021-2022
# ============================================================
PDFS = {
    2021: "C:/Users/hp elitebook/Projet_Data_ing/Bilans et comptes de résultat des banques 2021.pdf",
    2022: "C:/Users/hp elitebook/Projet_Data_ing/Bilans et comptes de résultat des banques 2022.pdf",
}

tous_resultats = []

for annee, pdf_path in PDFS.items():
    print(f"\n📂 Traitement PDF {annee}...")
    vus = set()
    with pdfplumber.open(pdf_path) as pdf:
        for i in range(len(pdf.pages)):
            text = pdf.pages[i].extract_text()
            if not text:
                continue
            if ("SENEGAL" in text.upper()
                and "comptes de r" in text.lower()
                and "PRODUITS/CHARGES" in text.upper()
                and ("CORIS" not in text.upper() or "CBI" in text.upper())):
                sigle_pdf = None
                for sigle in sorted(BANQUES_SENEGAL.keys(), key=len, reverse=True):
                    if sigle.upper() in text.upper():
                        sigle_pdf = sigle
                        break
                if sigle_pdf:
                    sigle_base = BANQUES_SENEGAL[sigle_pdf]
                    cle = f"{sigle_base}_{annee}"
                    if cle not in vus:
                        data = extraire_compte_resultat(pdf.pages[i], annee)
                        if data:
                            data['Sigle'] = sigle_base
                            data['ANNEE'] = annee
                            data['Goupe_Bancaire'] = None
                            tous_resultats.append(data)
                            vus.add(cle)
                            print(f"  ✅ {sigle_base} {annee}")

df_pdf = pd.DataFrame(tous_resultats)
print(f"\n✅ Total extrait : {len(df_pdf)} lignes")
print(df_pdf[['Sigle', 'ANNEE', 'PRODUIT.NET.BANCAIRE', 'RESULTAT.NET']].to_string())
```
